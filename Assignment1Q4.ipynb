```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):       # This creates a function returning the value of F(X) for a given x value
  y = 100-2*x**2+3*x     #This is the function that needs to be maximized according to the problem
  return(y)               #value of F(X) when given X

def maximization (a,b,N,f,showplot=True):    #defining the maximization function, including the previous funciton as an argument
  x=np.linspace(a,b,N)         #defining the values x can fall between and at which intervals
  y = f(x)                     #telling the code what value should take the y values on the scatterplot
  plt.scatter(x,y)             # generates the scatterplot
  max= np.max(y)
  maxy = np.where(y==max)
  maximizer = x[maxy]
  critvals = {"Max":maxy,"Maximizer":maximizer}
  print (critvals)

print(maximization(-1,1,3,f))
print(maximization(-1,1,10,f))
print(maximization(-1,1,100,f))
print(maximization(-1,1,1000,f))
print(maximization(-1,1,5000,f))
```
